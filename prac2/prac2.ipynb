2\. _Social networking_ Please look at the graph below of social relationships. You have a mango farm, you need to find a mango buyer and sell him your mango that you have grown (just a small farm). To this end, our principle is to use our social network of friends separated by one degree, and then second degree friends (friends of friends), third degree friends (friends of friends of friends).

Use the depth first algorithm to search your network until you find a mango seller. 

We make an assumption that if the person's name ends with m, then it means the person must be the mango seller!


<img src="img/p3.png" width="500" height="400" >

```python
class Graph():
    def __init__(self, nodes, sides):
        self.nodes = {} #key是结点， value是与key相连的结点
        self.side = []
        for node in nodes:
            for side in sides:
                u,v = side
                if node == u:
                    self.side.append(v)
                elif node == v:
                    self. side.append(u)
            self.nodes[node] =self.side
            self.side = []
            
                        
    def DFS (self,node_initial):
        
        queue,order=[],[]#queue用来存放需遍历的数据；order用来存放具体的访问路径
        queue.append(node_initial)
        while (queue):
            v = queue.pop()
            order.append(v)
            for w in self.nodes[v]:
                if w not in order and w not in queue:#即w没被访问
                    queue.append(w)
        return order

def main():
    nodes = ['you','bob','alice','claire','anuj','peggy','jonny','thom']
    sides = [('you','bob'),
            ('you','alice'),
            ('you','claire'),
            ('bob','anuj'),
            ('bob','peggy'),
            ('alice','peggy'),
            ('claire','jonny'),
            ('claire','thom')]
    G = Graph(nodes,sides)
    order_lists = G.DFS('you')
    for order_list in order_lists:
        if 'm' in str(order_list):
            print (order_list)
    
if __name__ == "__main__":
    main()
```

3\. Depth First traverse：Use depth first traversal to traverse all the DLL files on the C disk and output the file names.

Update your code to search for a particular filename.

What protocol did you use to select the next node to expand on the fringe?

```python
## 遍历c盘，将所有含.dll的文件找出
import os       
import fnmatch
 
def DFS(): 
  f = open('DFS_list.txt', 'w') #将目标文件写入文件DFS_list.txt
  for root, dirs, files in os.walk('c:\\'):
    for subFile in files:
      if fnmatch.fnmatch(subFile, '*.dll'):
        f.write(os.path.join(root, subFile))
        f.write('\n')
  f.close()
  print ('done...') #用来表明运行完成

def PARTICULAR(target_file): 
  for root, dirs, files in os.walk('c:\\'):
    for subFile in files:
      if fnmatch.fnmatch(subFile, target_file):
        print(os.path.join(root, subFile))
   
DFS()
PARTICULAR('$IU3SEWF.dll') 
```

Breadth First traverse：Use breadth traversal to traverse all the DLL files on the C disk and output the filename.

Update your code to search for a particular filename.

```python
## write your answer here!
import os       
import fnmatch
 
def BFS(): 
  f = open('BFS_list.txt', 'w') #将目标文件写入文件BFS_list.txt
  for root, dirs, files in os.walk('c:\\',topdown = True): #topdpwn 为True,优先遍历top 目录
    for subFile in files:
      if fnmatch.fnmatch(subFile, '*.dll'):
        f.write(os.path.join(root, subFile))
        f.write('\n')
  f.close()
  print ('done...') #用来表明运行完成

def PARTICULAR(target_file): 
  for root, dirs, files in os.walk('c:\\',topdown = True):
    for subFile in files:
      if fnmatch.fnmatch(subFile, target_file):
        print(os.path.join(root, subFile))
   
BFS()
PARTICULAR('$IU3SEWF.dll') 
```

5\. Uniform cost search

Consider a state space where the start state is 2 and each state k has three successors: numbers 2k, 2k+1, 2k+2. The cost from state k to each respective child is k, ground(k/2), k+2.

can you implement a uniform-cost-search algorithm with python. The goal state is number 85.

生成state是85的最小cost是多少

```python
import math

goal = 85
cost = [-1] * (goal + 1) #包含86个-1的列表
cost[2] = 0

for k in range(2, math.floor(goal/2) + 1): #因为85的k定多是42
    for successor, cos in [(2 * k, k), (2 *k + 1, math.floor(k/2)), (2 * k + 2, k + 2)]:
        if successor > goal:
            break
        if cost[successor] == -1 or cost[successor] > cost[k] + cos:
            cost[successor] = cost[k] + cos

print (cost[goal])
```
